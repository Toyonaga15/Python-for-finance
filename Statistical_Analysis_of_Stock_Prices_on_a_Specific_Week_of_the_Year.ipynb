```python
# Importing necessary libraries
import yfinance as yf
import pandas as pd
```

```python
#Asking user to input the stock name and the start and end years of analysis
asset = input('Enter the stock name according to yahoo finance: ').strip().upper()
start_year = int(input('Enter the starting year of analysis: '))
end_year = int(input('Enter the ending year of analysis: '))
week_number = int(input(f'Which week of the year do you want to analyze? (ex: 0, 1, 2, 3, 4, 5): '))

data = yf.download(asset, start=f'{start_year}-01-01', end=f'{end_year}-12-31', interval='1wk')

data_yf = pd.DataFrame(data)
```

```python
# Creating a dataframe and adding columns with percentage change

data_yf = pd.DataFrame(data)
data_yf['Closing %'] = data_yf['Close'].pct_change()*100
data_yf['Opening %'] = ((data_yf['Open']-data_yf['Close'].shift(1))/data_yf['Close'])*100
data_yf['Max %'] = ((data_yf['High']-data_yf['Close'].shift(1))/data_yf['Close'])*100
data_yf['Min %'] = ((data_yf['Low']-data_yf['Close'].shift(1))/data_yf['Close'])*100

# Changing the index to date
data_yf.index = pd.to_datetime(data_yf.index.date)
```

```python
# Selecting only the data of the selected week number
week_data = data[data.index.week == week_number]
week_data = pd.DataFrame(week_data)

df = week_data
df
```

```python
# Creating an empty list to store the result of closing percentage change (positive or negative)
closing_result = []

# Loop to check if the closing percentage change is positive or negative
for i in df['Closing %']:
  if i > 0:
    closing_result.append('Positive')
  else:
    closing_result.append('Negative')

# Adding a new column to the dataframe with the closing result
df['Closing Result'] = closing_result

# Creating an empty list to store the result of opening percentage change (positive or negative)
opening_result = []

# Loop to check if the opening percentage change is positive or negative

for i in df['Opening %']:
  if i > 0:
    opening_result.append('Positive')
  else:
    opening_result.append('Negative')

# Adding a new column to the dataframe with the opening result
df['Opening Result'] = opening_result

# Counting the number of positive closing results and dividing by the total number of data points to get the probability
positive_closing_count = 0
for i in df['Closing Result']:
  if i =='Positive':
    positive_closing_count += 1
    prob_closing = (positive_closing_count/len(df['Closing Result']))*100
    


# Dividing the count of positive closing results by the total number of data points to get the probability of a positive closing result
prob_closing = (positive_closing_count / len(df['Closing Result'])) * 100

# Counting the number of positive opening results and dividing by the total number of data points to get the probability

positive_opening_count = 0
for i in df['Opening Result']:
  if i =='Positive':
    positive_opening_count += 1
    prob_opening = (positive_opening_count / len(df['Opening Result'])) * 100

# Calculating the mean of the max and min percentage changes
mean_max = sum(df['Max %']) / len(df['Max %'])
mean_min = sum(df['Min %']) / len(df['Min %'])

# Printing the final result
print(f'''Statistically, the stock {asset} opened {round(prob_opening,2)}% of the times in high and closed
{round(prob_closing,2)}% of the times in high in the selected dates and weeks.''')
print(f'The mean of the max percentage changes is {round(mean_max,2)}% and the mean of the min percentage changes is {round(mean_min,2)}%')

```
