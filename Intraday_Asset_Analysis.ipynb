```python
!pip install yfinance
import pandas as pd
import datetime
import yfinance as yf
from dateutil.relativedelta import relativedelta
from dateutil.relativedelta import relativedelta,MO,TU,WE,TH,FR,SA,SU
```

```python
start_year = int(input('Enter the initial year of the analysis: '))
end_year = int(input('Enter the final year of the analysis: '))
mounth = int(input('Which month do you want to analyze? (ex: 1, 2, 3...): '))
day = input('Which day of the week do you want to analyze? (ex: Mo, Tu, We...): ').strip().upper()
week = int(input(f'Which {day} of the month do you want to analyze? (ex: 1, 2, 3, 4, 5): '))
```

```python
# Get the asset name and download the data from Yahoo Finance
asset = input('Enter the asset name according to Yahoo Finance: ').strip().upper()
data_yf = yf.download(asset, start=f'{start_year}-01-01', end=f'{end_year}-12-31', interval='1d')
```

```python
# Add new columns with the percentage change of the opening, closing, high, and low values
data_yf['Closing in %']= data_yf['Close'].pct_change()*100
data_yf['Opening in %'] = ((data_yf['Open']-data_yf['Close'].shift(1))/data_yf['Close'])*100
data_yf['Max in %'] = ((data_yf['High']-data_yf['Close'].shift(1))/data_yf['Close'])*100
data_yf['Min in %'] = ((data_yf['Low']-data_yf['Close'].shift(1))/data_yf['Close'])*100

# Convert the index to a datetime format
data_yf.index = pd.to_datetime(data_yf.index.date)
data_yf.tail()
```

```python
# Convert the day input to the corresponding date format
if day == 'MO':
  day = MO
elif day == 'TU':
  day = TU
elif day == 'WE':
  day = WE
elif day == 'TH':
  day = TH
elif day == 'FR':
  day = FR
elif day == 'SA':
  day = SA
elif day == 'SU':
  day = SU

i = 0
dates=[]

# Get the dates of the selected day of the week in the selected month
for i in range (start_year,end_year+1):
  date = datetime.date(i, 1, 1)
  i=+1
  answer = date + relativedelta(months= mounth-1, weekday= day(week))
  dates.append(answer)
  print(str(f'{answer}.'),end=" ")
```

```python
# Filter the dataframe to only include the selected dates
df = data_yf.loc[data_yf.index.isin(dates)]
df
```

```python
# Create a new column indicating whether the closing value was positive or negative
resultC = []

for i in df['Closing in %']:
  if i > 0:
    resultC.append('Positive')
  else:
    resultC.append('Negative')

df['Result C'] = resultC

# Create a new column indicating whether the opening value was positive or negative
resultO = []

for i in df['Opening in %']:
  if i > 0:
    resultO.append('Positive')
  else:
    resultO.append('Negative')

df['Result O'] = resultO

# Count the number of times the closing value was positive
c = 0
for i in df['Result C']:
  if i =='Positive':
    c=c+1
statC = (c/len(df['Result C']))*100

# Count the number of times the closing value was positive
d = 0
for i in df['Result O']:
  if i =='Positive':
    d=d+1
statO = (d/len(df['Result O']))*100

# Calculate the average of the maximum and minimum values
mmax = sum(df['Max in %'])/len(df['Max in %'])
mmin = sum(df['Min in %'])/len(df['Min in %'])

# Print the results
print(f'''Statistically the asset {asset} opened {round(statO,2)}% of the times in high and closed
{round(statC,2)}% of the times in high in the selected dates.''')
print(f'The average of the maximums is {round(mmax,2)}% and of the minimums {round(mmin,2)}%\n')
```
